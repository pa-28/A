## PART 1: BASIC CSP EXAMPLE

Use when question asks:
"Create variables and apply simple constraints"
Example: a, b, c with different values

````python
def basic_csp():
    print("\n===== BASIC CSP EXAMPLE =====")

    model = cp_model.CpModel()

    # Create variables with domain 1 to 3
    a = model.NewIntVar(1, 3, 'a')
    b = model.NewIntVar(1, 3, 'b')
    c = model.NewIntVar(1, 3, 'c')

    # Constraint: all values must be different
    model.Add(a != b)
    model.Add(b != c)
    model.Add(a != c)

    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        print("a =", solver.Value(a))
        print("b =", solver.Value(b))
        print("c =", solver.Value(c))
    else:
        print("No solution found")
````

## PART 2: EMPLOYEE SHIFT ASSIGNMENT

Use when question asks:
"Assign shifts with constraints"

````python
def employee_shift():
    print("\n===== EMPLOYEE SHIFT ASSIGNMENT =====")

    model = cp_model.CpModel()

    # Shift values: 0, 1, 2
    alice = model.NewIntVar(0, 2, 'Alice')
    bob = model.NewIntVar(0, 2, 'Bob')
    charlie = model.NewIntVar(0, 2, 'Charlie')

    # Constraints
    model.Add(alice != bob)   # Alice and Bob cannot have same shift
    model.Add(bob != 0)       # Bob cannot take shift 0

    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        print("Alice Shift =", solver.Value(alice))
        print("Bob Shift =", solver.Value(bob))
        print("Charlie Shift =", solver.Value(charlie))
    else:
        print("No valid assignment")
````

## PART 3: N-QUEENS PROBLEM

Use when question asks:
"Solve 8 queens using CSP"

````python
def n_queens():
    print("\n===== N-QUEENS PROBLEM =====")

    n = 8
    model = cp_model.CpModel()

    # queens[i] = row position of queen in column i
    queens = [model.NewIntVar(0, n - 1, f'q{i}') for i in range(n)]

    # No two queens in same row
    model.AddAllDifferent(queens)

    # Diagonal constraints
    diag1 = []
    diag2 = []

    for i in range(n):
        d1 = model.NewIntVar(0, 2 * n, f'd1_{i}')
        d2 = model.NewIntVar(-n, n, f'd2_{i}')

        model.Add(d1 == queens[i] + i)
        model.Add(d2 == queens[i] - i)

        diag1.append(d1)
        diag2.append(d2)

    model.AddAllDifferent(diag1)
    model.AddAllDifferent(diag2)

    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        for i in range(n):
            print(f"Column {i} -> Row {solver.Value(queens[i])}")
    else:
        print("No solution")
````

## PART 4: ALLOWED / FORBIDDEN ASSIGNMENTS

Use when question asks:
"Use AddAllowedAssignments or AddForbiddenAssignments"

````python
def allowed_forbidden():
    print("\n===== ALLOWED / FORBIDDEN ASSIGNMENTS =====")

    model = cp_model.CpModel()

    x = model.NewIntVar(1, 3, 'x')
    y = model.NewIntVar(1, 3, 'y')

    # Only these combinations are allowed
    allowed = [(1, 2), (2, 3), (3, 1)]
    model.AddAllowedAssignments([x, y], allowed)

    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        print("x =", solver.Value(x))
        print("y =", solver.Value(y))
    else:
        print("No solution")
````

````python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []
        self.minmax_value = None
````

## PART 3: MINIMAX ALGORITHM

Use when question asks:
"Implement Minimax"

````python
def minimax(node, depth, maximizing_player=True):
    if depth == 0 or not node.children:
        return node.value

    if maximizing_player:
        best_value = -math.inf
        for child in node.children:
            child_value = minimax(child, depth - 1, False)
            best_value = max(best_value, child_value)
        node.minmax_value = best_value
        return best_value

    else:
        best_value = math.inf
        for child in node.children:
            child_value = minimax(child, depth - 1, True)
            best_value = min(best_value, child_value)
        node.minmax_value = best_value
        return best_value
````

## PART 4: AGENT + ENVIRONMENT FOR MINIMAX

Use when teacher asks PEAS / Agent Environment model

````python
class MinimaxAgent:
    def __init__(self, depth):
        self.depth = depth

    def act(self, node, environment):
        return environment.compute_minimax(node, self.depth)


class Environment:
    def __init__(self, tree):
        self.tree = tree
        self.computed_nodes = []

    def compute_minimax(self, node, depth, maximizing_player=True):
        if depth == 0 or not node.children:
            self.computed_nodes.append(node.value)
            return node.value

        if maximizing_player:
            best_value = -math.inf
            for child in node.children:
                best_value = max(
                    best_value,
                    self.compute_minimax(child, depth - 1, False)
                )
            node.minmax_value = best_value
            return best_value

        else:
            best_value = math.inf
            for child in node.children:
                best_value = min(
                    best_value,
                    self.compute_minimax(child, depth - 1, True)
                )
            node.minmax_value = best_value
            return best_value
````

## PART 5: ALPHA-BETA PRUNING

Use when question asks:
"Implement Alpha-Beta pruning"

````python
def alpha_beta(node, depth, alpha, beta, maximizing_player=True):
    if depth == 0 or not node.children:
        return node.value

    if maximizing_player:
        best_value = -math.inf
        for child in node.children:
            best_value = max(
                best_value,
                alpha_beta(child, depth - 1, alpha, beta, False)
            )
            alpha = max(alpha, best_value)

            if beta <= alpha:
                print("Pruned node:", child.value)
                break

        node.minmax_value = best_value
        return best_value

    else:
        best_value = math.inf
        for child in node.children:
            best_value = min(
                best_value,
                alpha_beta(child, depth - 1, alpha, beta, True)
            )
            beta = min(beta, best_value)

            if beta <= alpha:
                print("Pruned node:", child.value)
                break

        node.minmax_value = best_value
        return best_value
````

## PART 1: CONDITIONAL PROBABILITY

Use when question asks:
"Find P(King | Face Card) using simulation"

````python
import random


def conditional_probability_cards():
    print("\n===== CONDITIONAL PROBABILITY =====")

    suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
    ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10',
             'Jack', 'Queen', 'King', 'Ace']

    # Create full deck
    deck = [(rank, suit) for suit in suits for rank in ranks]

    # Only face cards
    face_cards = [card for card in deck if card[0] in ['Jack', 'Queen', 'King']]

    trials = 100000
    king_count = 0

    for _ in range(trials):
        card = random.choice(face_cards)
        if card[0] == 'King':
            king_count += 1

    simulated_probability = king_count / trials
    theoretical_probability = 4 / 12

    print("Theoretical P(King | Face Card) =", theoretical_probability)
    print("Simulated P(King | Face Card) =", round(simulated_probability, 3))
````

## PART 2: BAYESIAN NETWORK (MONTY HALL)

Use when question asks:
"Create Bayesian Network using pgmpy"

````python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination


def monty_hall_bn():
    print("\n===== MONTY HALL BAYESIAN NETWORK =====")

    model = DiscreteBayesianNetwork([
        ('Guest', 'Monty'),
        ('Prize', 'Monty')
    ])

    cpd_guest = TabularCPD(
        variable='Guest',
        variable_card=3,
        values=[[1/3], [1/3], [1/3]],
        state_names={'Guest': [1, 2, 3]}
    )

    cpd_prize = TabularCPD(
        variable='Prize',
        variable_card=3,
        values=[[1/3], [1/3], [1/3]],
        state_names={'Prize': [1, 2, 3]}
    )

    # Simplified CPD for Monty
    cpd_monty = TabularCPD(
        variable='Monty',
        variable_card=3,
        values=[
            [0, 0, 0, 0, 0.5, 1, 0, 1, 0.5],
            [0.5, 0, 1, 0, 0, 0, 1, 0, 0.5],
            [0.5, 1, 0, 1, 0.5, 0, 0, 0, 0]
        ],
        evidence=['Guest', 'Prize'],
        evidence_card=[3, 3],
        state_names={
            'Monty': [1, 2, 3],
            'Guest': [1, 2, 3],
            'Prize': [1, 2, 3]
        }
    )

    model.add_cpds(cpd_guest, cpd_prize, cpd_monty)
    print("Model Valid:", model.check_model())

    inference = VariableElimination(model)
    result = inference.query(variables=['Prize'])
    print(result)
````

## PART 3: BURGLARY ALARM PROBLEM

Use when question asks:
"Alarm Bayesian Network"

````python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Step 1: Define the structure of the Bayesian Network
model = DiscreteBayesianNetwork([
    ('Burglary', 'Alarm'),
    ('Earthquake', 'Alarm'),
    ('Alarm', 'JohnCalls'),
    ('Alarm', 'MaryCalls')
])

# Step 2: Define the CPDs (Conditional Probability Distributions)

# P(Burglary)
cpd_burglary = TabularCPD(
    variable='Burglary',
    variable_card=2,
    values=[[0.999], [0.001]]
)

# P(Earthquake)
cpd_earthquake = TabularCPD(
    variable='Earthquake',
    variable_card=2,
    values=[[0.998], [0.002]]
)

# P(Alarm | Burglary, Earthquake)
cpd_alarm = TabularCPD(
    variable='Alarm',
    variable_card=2,
    values=[
        [0.999, 0.71, 0.06, 0.05],  # Alarm = False
        [0.001, 0.29, 0.94, 0.95]   # Alarm = True
    ],
    evidence=['Burglary', 'Earthquake'],
    evidence_card=[2, 2]
)

# P(JohnCalls | Alarm)
cpd_john = TabularCPD(
    variable='JohnCalls',
    variable_card=2,
    values=[
        [0.3, 0.9],  # JohnCalls = False
        [0.7, 0.1]   # JohnCalls = True
    ],
    evidence=['Alarm'],
    evidence_card=[2]
)

# P(MaryCalls | Alarm)
cpd_mary = TabularCPD(
    variable='MaryCalls',
    variable_card=2,
    values=[
        [0.2, 0.99],  # MaryCalls = False
        [0.8, 0.01]   # MaryCalls = True
    ],
    evidence=['Alarm'],
    evidence_card=[2]
)

# Step 3: Add CPDs to the model
model.add_cpds(cpd_burglary, cpd_earthquake, cpd_alarm, cpd_john, cpd_mary)

# Step 4: Verify the model
assert model.check_model(), "Model is incorrect"

# Step 5: Perform inference
inference = VariableElimination(model)

# Query
result = inference.query(
    variables=['Burglary'],
    evidence={'JohnCalls': 1, 'MaryCalls': 1}
)

print(result)
````

## PART 4: MARKOV PROCESS

Use when question asks:
"Simulate Markov Chain"

````python
import numpy as np


def markov_process():
    print("\n===== MARKOV PROCESS =====")

    states = ['Red', 'Blue']

    transition_matrix = np.array([
        [0.5, 0.5],
        [0.5, 0.5]
    ])

    def simulate(initial_state, steps):
        current_state = initial_state
        sequence = [current_state]

        for _ in range(steps):
            if current_state == 'Red':
                next_state = np.random.choice(states, p=transition_matrix[0])
            else:
                next_state = np.random.choice(states, p=transition_matrix[1])

            sequence.append(next_state)
            current_state = next_state

        return sequence

    result = simulate('Red', 10)
    print("State Sequence:", result)
````

````python
# ML Cheatsheet — Full DS Cycle

---

## IMPORTS (always at the top)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.metrics import (
    accuracy_score, mean_squared_error, mean_absolute_error,
    r2_score, classification_report, confusion_matrix, roc_curve, auc
)
```

---

## STEP 1 — DATA LOADING & EXPLORATION

```python
df = pd.read_csv('file.csv')       # Load CSV
df = pd.read_excel('file.xlsx')    # Load Excel
df = pd.read_json('file.json')     # Load JSON

df.head()                          # First 5 rows
df.tail()                          # Last 5 rows
df.shape                           # (rows, columns)
df.info()                          # Column names, types, non-null counts
df.describe()                      # Mean, std, min, max per column
df.dtypes                          # Data types
df.columns                         # Column names

df.isnull().sum()                  # Count NaN per column
(df.isnull().sum() / len(df)) * 100  # % missing per column

df['col'].value_counts()           # Frequency of each category
df.select_dtypes(include='object') # Only categorical columns
df.select_dtypes(include='number') # Only numerical columns
df.skew()                          # Skewness (0=normal, +/-=skewed)
df.corr()                          # Correlation matrix
```

---

## STEP 2 — PREPROCESSING

### Handle Missing Values
```python
df.dropna()                        # Drop rows with ANY NaN
df.dropna(subset=['col'])          # Drop rows where specific col is NaN
df['col'].fillna(df['col'].mean()) # Fill NaN with mean
df['col'].fillna(df['col'].median()) # Fill NaN with median
df['col'].fillna(df['col'].mode()[0]) # Fill NaN with mode (for categorical)
```

### Encoding Categorical Variables
```python
# Label Encoding (for binary: Yes/No, Male/Female)
le = LabelEncoder()
df['col'] = le.fit_transform(df['col'])

# One-Hot Encoding (for multi-class: Department, JobRole)
df = pd.get_dummies(df, columns=['col'], drop_first=True)
```

### Separate Features and Target
```python
X = df.drop('target', axis=1)     # All columns except target
y = df['target']                   # Target column only
```

### Train-Test Split
```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 80/20 split
    random_state=42      # Reproducibility
)
# Other splits: test_size=0.3 (70/30), test_size=0.25 (75/25)
```

### Feature Scaling
```python
# StandardScaler: mean=0, std=1 (use for most algorithms)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # fit AND transform
X_test_scaled = scaler.transform(X_test)        # transform ONLY

# MinMaxScaler: range [0, 1]
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

> **Rule:** Always `fit_transform` on TRAIN, `transform` only on TEST.  
> Never fit on test data — that's cheating.

### Outlier Detection
```python
# IQR Method
Q1 = df['col'].quantile(0.25)
Q3 = df['col'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['col'] < lower) | (df['col'] > upper)]

# Z-Score Method
from scipy import stats
z_scores = np.abs(stats.zscore(df.select_dtypes(include='number')))
outlier_rows = (z_scores > 3).any(axis=1)
df_clean = df[~outlier_rows]
```

### Outlier Treatment
```python
# Cap outliers (Winsorizing)
df['col'] = df['col'].clip(lower=lower, upper=upper)

# Drop outliers
df = df[~((df['col'] < lower) | (df['col'] > upper))]
```

### Feature Engineering
```python
# Create new features from existing ones
df['IncomePerYear'] = df['MonthlyIncome'] * 12
df['SatisfactionScore'] = (df['JobSatisfaction'] + df['EnvironmentSatisfaction']) / 2
df['LongTenure'] = (df['YearsAtCompany'] > 10).astype(int)  # Binary flag
```

### Class Imbalance
```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Oversampling (adds minority class samples)
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_train, y_train)

# Undersampling (removes majority class samples)
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)
```

---

## STEP 3 — MODEL TRAINING

> The template is ALWAYS the same. Just swap the model.

```python
model = LogisticRegression()          # swap this line
model.fit(X_train_scaled, y_train)    # train
y_pred = model.predict(X_test_scaled) # predict
```

### Available Models

```python
# Classification
model = LogisticRegression()
model = DecisionTreeClassifier(max_depth=5, random_state=42)
model = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
model = GaussianNB()
model = SVC(kernel='rbf')

# Regression
model = LinearRegression()
model = DecisionTreeRegressor(max_depth=5, random_state=42)
```

### When to use what?
| Model | Use When |
|-------|----------|
| Linear Regression | Predict continuous number (price, salary) |
| Logistic Regression | Binary/multi-class classification |
| Decision Tree | Classification or regression, interpretable |
| KNN | Small datasets, no training phase |
| Naive Bayes | Text classification, fast |
| SVM | High-dimensional data, good accuracy |
| K-Means | Clustering (unsupervised, no labels) |

---

## STEP 4 — EVALUATION

### For Classification
```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

print(classification_report(y_test, y_pred))
# Shows: precision, recall, f1-score per class

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

**Confusion Matrix:**
```
                 Predicted YES   Predicted NO
Actual YES    |  True Positive  | False Negative |
Actual NO     |  False Positive | True Negative  |
```
- **Accuracy** = (TP + TN) / Total
- **Precision** = TP / (TP + FP) — of those I said YES, how many were right?
- **Recall** = TP / (TP + FN) — of all actual YES, how many did I catch?
- **F1** = 2 × (Precision × Recall) / (Precision + Recall)

### For Regression
```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²: {r2:.4f}")
```

- **MAE** — average absolute error (easy to interpret)
- **MSE** — penalizes large errors more (squared)
- **RMSE** — same units as target, most commonly reported
- **R²** — how much variance is explained (1.0 = perfect)

### ROC Curve (Binary Classification)
```python
y_prob = model.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {auc_score:.2f}')
plt.plot([0, 1], [0, 1], 'k--')  # Random classifier line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

---

## CROSS VALIDATION

```python
# K-Fold Cross Validation
from sklearn.model_selection import cross_val_score

model = LogisticRegression()
scores = cross_val_score(model, X, y, cv=10)  # 10-fold
print(f"Average Accuracy: {scores.mean():.4f}")
print(f"Std Dev: {scores.std():.4f}")

# Manual K-Fold
kf = KFold(n_splits=10, shuffle=True, random_state=42)
for train_idx, test_idx in kf.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
```

---

## STEP 5 — VISUALIZATION

```python
# 1. Histogram
df['col'].hist(bins=30)
plt.title('Distribution of col')
plt.show()

# 2. Bar chart (categorical counts)
df['col'].value_counts().plot(kind='bar')
plt.title('Value Counts')
plt.show()

# 3. Boxplot (outliers + spread)
df.boxplot(column='col', by='target')
plt.show()

# 4. Confusion Matrix Heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# 5. Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 6. Scatter plot
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # ideal line
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()

# 7. Pairplot
sns.pairplot(df, hue='target')
plt.show()

# 8. KDE plot (distribution with curve)
sns.kdeplot(df['col'], shade=True)
plt.show()
```

---

## UNSUPERVISED LEARNING — K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Prepare features (drop non-numeric or ID columns)
X = df.drop(['id_col', 'categorical_col'], axis=1)

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method — find optimal K
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.show()

# Train with optimal K (pick from elbow graph, e.g. K=5)
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_pred = kmeans.fit_predict(X_scaled)

# Add cluster labels back to dataframe
df['Cluster'] = y_pred

# Visualize clusters (pick 2 features for axes)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='red', marker='X', label='Centroids')
plt.title('K-Means Clusters')
plt.legend()
plt.show()
```

**K-Means Steps (theory):**
1. Choose K (number of clusters)
2. Randomly place K centroids
3. Assign each point to nearest centroid
4. Recalculate centroids as mean of assigned points
5. Repeat 3-4 until no reassignment happens

**WCSS** = Within Cluster Sum of Squares (lower = tighter clusters)

---

## OVERFITTING vs UNDERFITTING

| | Train Accuracy | Test Accuracy |
|--|--|--|
| **Underfitting** | Low | Low |
| **Overfitting** | High | Low |
| **Good fit** | High | High (similar) |

**Fix overfitting:** Reduce model complexity (`max_depth`), use cross-validation, get more data  
**Fix underfitting:** Use more complex model, add features, remove regularization

---

## QUICK REFERENCE

```python
# Describe one column
df['col'].describe()

# Check unique values
df['col'].unique()
df['col'].nunique()

# Group comparison
df.groupby('target')[['col1', 'col2']].mean()

# Crosstab (categorical vs categorical)
pd.crosstab(df['col1'], df['col2'])

# Drop column
df.drop('col', axis=1, inplace=True)

# Rename column
df.rename(columns={'old': 'new'}, inplace=True)

# Filter rows
df[df['col'] > 5]
df[(df['col1'] > 5) & (df['col2'] == 'Yes')]
```
````

````python
# ----------------- Part 1: Data Preprocessing & EDA --------------------------

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

sns.set_style('whitegrid')

# Load the California Housing dataset
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target  # target column

print('Shape of dataset:', df.shape)
print(df.head())

# Dataset structure
df.info()

# Summary statistics
print(df.describe())

# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())

# Scatter plot: Median Income vs House Value
plt.figure(figsize=(8, 5))
plt.scatter(df['MedInc'], df['MedHouseVal'], alpha=0.3, color='teal')
plt.xlabel('Median Income (MedInc)')
plt.ylabel('Median House Value')
plt.title('Median Income vs Median House Value')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - California Housing')
plt.show()

# Distribution of target variable
plt.figure(figsize=(8, 4))
sns.histplot(df['MedHouseVal'], bins=40, kde=True, color='purple')
plt.title('Distribution of Median House Value')
plt.xlabel('MedHouseVal')
plt.show()

# EDA Insights:
# - Dataset has 20,640 rows and 9 columns; no missing values.
# - MedInc has the strongest positive correlation with house value.
# - Target variable is right-skewed and capped at 5.0.

# Separate features (X) and target (y)
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)


# ----------------- Part 2: Model Training ------------------------------------

# Split into training and testing sets (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Model 1: Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

# Model 2: Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42, max_depth=10)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)


# ----------------- Part 3: Model Evaluation & Visualization ------------------

mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr  = r2_score(y_test, y_pred_lr)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt  = r2_score(y_test, y_pred_dt)

results = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree'],
    'MSE'  : [mse_lr, mse_dt],
    'R2'   : [r2_lr, r2_dt]
})
print(results)

# Actual vs Predicted plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(y_test, y_pred_lr, alpha=0.3, color='steelblue')
axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
axes[0].set_xlabel('Actual')
axes[0].set_ylabel('Predicted')
axes[0].set_title('Linear Regression: Actual vs Predicted')

axes[1].scatter(y_test, y_pred_dt, alpha=0.3, color='darkorange')
axes[1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
axes[1].set_xlabel('Actual')
axes[1].set_ylabel('Predicted')
axes[1].set_title('Decision Tree: Actual vs Predicted')

plt.tight_layout()
plt.show()

# Bar chart comparison of R^2 scores
plt.figure(figsize=(7, 4))
sns.barplot(x='Model', y='R2', data=results, palette=['steelblue', 'darkorange'])
plt.title('Model R2 Score Comparison')
plt.ylim(0, 1)
for i, v in enumerate(results['R2']):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', fontweight='bold')
plt.show()

# Conclusion:
# Decision Tree Regressor (R2 ~ 0.70) outperforms Linear Regression (R2 ~ 0.58)
# because house prices have non-linear relationships with the features.
````

````python
# ----------------- Part 1: Data Preprocessing & EDA --------------------------

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')
print('Shape:', df.shape)
print(df.head())

# Dataset structure
df.info()

# Summary statistics
print(df.describe(include='all'))

# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())

# Survival count plot
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=df, palette=['salmon', 'seagreen'])
plt.title('Survival Count (0 = Died, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.show()

# Survival by gender
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='survived', data=df, palette=['salmon', 'seagreen'])
plt.title('Survival by Gender')
plt.show()

# Drop columns that are duplicates or not useful for modelling
# 'deck' has too many missing values; 'alive' duplicates the target;
# 'who', 'adult_male', 'class', 'embark_town', 'alone' are derived from others
df = df.drop(columns=['deck', 'alive', 'who', 'adult_male',
                      'class', 'embark_town', 'alone'])

# Handle missing values
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

print('Missing values after handling:')
print(df.isnull().sum())

# Encode categorical variables
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)
df = df.astype({col: 'int' for col in df.select_dtypes('bool').columns})

# Correlation heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Titanic')
plt.show()

# EDA Insights:
# - About 38% of passengers survived.
# - Females survived at much higher rates than males.
# - Higher class passengers (pclass=1) had better chances of survival.
# - Filled missing 'age' with median and 'embarked' with mode.

# Separate features (X) and target (y)
X = df.drop('survived', axis=1)
y = df['survived']

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)


# ----------------- Part 2: Model Training ------------------------------------

# Split (stratified to keep class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Model 1: Logistic Regression
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)

# Model 2: Decision Tree Classifier
dt_model = DecisionTreeClassifier(random_state=42, max_depth=5)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)


# ----------------- Part 3: Model Evaluation & Visualization ------------------

acc_log = accuracy_score(y_test, y_pred_log)
acc_dt  = accuracy_score(y_test, y_pred_dt)

print(f'Logistic Regression Accuracy : {acc_log:.4f}')
print(f'Decision Tree Accuracy       : {acc_dt:.4f}')

# Detailed classification reports
print('--- Logistic Regression ---')
print(classification_report(y_test, y_pred_log))
print('--- Decision Tree ---')
print(classification_report(y_test, y_pred_dt))

# Confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

cm_log = confusion_matrix(y_test, y_pred_log)
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
axes[0].set_title('Logistic Regression - Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

cm_dt = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Oranges', ax=axes[1],
            xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
axes[1].set_title('Decision Tree - Confusion Matrix')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.tight_layout()
plt.show()

# Accuracy comparison bar chart
plt.figure(figsize=(7, 4))
models = ['Logistic Regression', 'Decision Tree']
accs   = [acc_log, acc_dt]
bars = plt.bar(models, accs, color=['steelblue', 'darkorange'])
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 1)
for bar, acc in zip(bars, accs):
    plt.text(bar.get_x() + bar.get_width()/2, acc + 0.01,
             f'{acc:.3f}', ha='center', fontweight='bold')
plt.show()

# Conclusion:
# Both models achieve ~0.78-0.82 accuracy. Logistic Regression is generally
# slightly more stable; Decision Tree's max_depth=5 prevents overfitting.
# Most important features: sex, pclass, fare.
````

## QUESTION 3: CLUSTERING TASK - IRIS DATASET (K-MEANS)

````python
# ----------------- Part 1: Data Preprocessing & EDA --------------------------

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]  # kept only for later comparison

print('Shape:', df.shape)
print(df.head())

# Dataset structure
df.info()

# Summary statistics
print(df.describe())

# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())

# Scatter plot: Sepal Length vs Petal Length
plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)',
                hue='species', data=df, palette='Set1', s=70)
plt.title('Sepal Length vs Petal Length (by species)')
plt.show()

# Correlation heatmap (numeric columns only)
plt.figure(figsize=(7, 5))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Iris')
plt.show()

# Pairplot
sns.pairplot(df, hue='species', palette='Set1')
plt.show()

# EDA Insights:
# - 150 rows, 4 numerical features, no missing values.
# - Petal length and petal width strongly correlated (r ~ 0.96).
# - Setosa is linearly separable; Versicolor and Virginica overlap slightly.

# Define feature set X (drop the species label - this is unsupervised learning)
X = df.drop('species', axis=1)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)


# ----------------- Part 2: Model Training ------------------------------------

# Compute silhouette scores for k = 2 to 10
silhouette_scores = []
k_values = range(2, 11)

for k in k_values:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
    print(f'k = {k}  ->  silhouette score = {score:.4f}')

# Plot silhouette scores
plt.figure(figsize=(8, 4))
plt.plot(list(k_values), silhouette_scores, marker='o', color='teal')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method for Optimal k')
plt.xticks(list(k_values))
plt.grid(True)
plt.show()

# Pick k with highest silhouette score
optimal_k = list(k_values)[int(np.argmax(silhouette_scores))]
print(f'Optimal number of clusters: k = {optimal_k}')

# Train final K-Means with optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)
df['cluster'] = cluster_labels


# ----------------- Part 3: Model Evaluation & Visualization ------------------

final_score = silhouette_score(X_scaled, cluster_labels)
print(f'Final Silhouette Score (k={optimal_k}): {final_score:.4f}')

# Scatter plot of clusters with centroids (Petal Length vs Petal Width)
plt.figure(figsize=(9, 6))
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'],
                hue=df['cluster'], palette='Set2', s=80, edgecolor='black')

# Convert centroids back to original scale for plotting
centroids_scaled = kmeans.cluster_centers_
centroids_original = scaler.inverse_transform(centroids_scaled)
centroids_df = pd.DataFrame(centroids_original, columns=X.columns)

plt.scatter(centroids_df['petal length (cm)'], centroids_df['petal width (cm)'],
            marker='X', s=300, c='red', edgecolor='black',
            label='Centroids', linewidth=2)

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title(f'K-Means Clusters (k={optimal_k}) with Centroids')
plt.legend()
plt.show()

# Optional: compare clusters with actual species
comparison = pd.crosstab(df['species'], df['cluster'],
                          rownames=['Actual Species'], colnames=['Cluster'])
print('Cluster vs Actual Species:')
print(comparison)

plt.figure(figsize=(7, 4))
sns.heatmap(comparison, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Cluster Assignments vs Actual Species')
plt.show()

# Conclusion:
# Silhouette suggests k=2 (~0.58) but k=3 (~0.46) matches the 3 actual species.
# Setosa is perfectly separated; Versicolor and Virginica overlap slightly.
````

````python
# PAPER A Q1 VERSION 1
"""
Alpha-Beta Pruning Implementation for Tic-Tac-Toe Endgame Tree

Tree structure (as per problem statement but with 9 leaves):
- Level 0 (root): MIN (opponent)
- Level 1: MAX (AI agent) nodes (3 nodes)
- Level 2: Leaf nodes (heuristic scores), 3 per MAX node

Leaf values (left to right): [3, 9, 1, 8, 5, 2, 7, 4, 6]
"""

import math

# Leaf values in order of traversal (left to right)
leaf_values = [3, 9, 1, 8, 5, 2, 7, 4, 6]
branching_factor = 3

# Global pruning log
pruned_nodes = []

def alpha_beta_min(node_depth, alpha, beta, node_path, leaf_index):
    """
    MIN node (opponent) at given depth.
    Returns (best_value, new_leaf_index, log_string)
    """
    global pruned_nodes
    if node_depth == 0:  # leaf level
        val = leaf_values[leaf_index]
        log = f"Leaf at path {node_path} = {val}\n"
        return val, leaf_index + 1, log
    
    value = math.inf
    log = ""
    for i in range(branching_factor):
        child_path = node_path + [i]
        # Recursively evaluate MAX child
        child_val, leaf_index, child_log = alpha_beta_max(node_depth - 1, alpha, beta, child_path, leaf_index)
        log += child_log
        if child_val < value:
            value = child_val
        beta = min(beta, value)
        # Pruning condition
        if value <= alpha:
            # Prune remaining branches
            pruned_nodes.append((child_path, "MIN node pruning"))
            log += f"Pruned at MIN node {child_path}: value={value} <= alpha={alpha}\n"
            break
    log += f"MIN node at path {node_path} returns {value}, alpha={alpha}, beta={beta}\n"
    return value, leaf_index, log

def alpha_beta_max(node_depth, alpha, beta, node_path, leaf_index):
    """
    MAX node (AI) at given depth.
    Returns (best_value, new_leaf_index, log_string)
    """
    if node_depth == 0:
        val = leaf_values[leaf_index]
        log = f"Leaf at path {node_path} = {val}\n"
        return val, leaf_index + 1, log
    
    value = -math.inf
    log = ""
    for i in range(branching_factor):
        child_path = node_path + [i]
        child_val, leaf_index, child_log = alpha_beta_min(node_depth - 1, alpha, beta, child_path, leaf_index)
        log += child_log
        if child_val > value:
            value = child_val
        alpha = max(alpha, value)
        if value >= beta:
            pruned_nodes.append((child_path, "MAX node pruning"))
            log += f"Pruned at MAX node {child_path}: value={value} >= beta={beta}\n"
            break
    log += f"MAX node at path {node_path} returns {value}, alpha={alpha}, beta={beta}\n"
    return value, leaf_index, log

def main():
    global pruned_nodes
    print("=== Alpha-Beta Pruning on Tic-Tac-Toe Endgame Tree ===\n")
    print("Tree: Root (MIN) -> 3 MAX nodes -> 3 leaves each.")
    print(f"Leaf values left->right: {leaf_values}\n")
    
    # Start at root MIN node, depth = 2 (two levels: MIN then MAX then leaf? Actually root depth 2 means 2 levels of internal nodes? Let's use depth = 2 because we have MIN at root, MAX at level1, leaves at level2. So depth=2 for MIN node means it has 2 more layers? Better: We define recursion depth as number of moves remaining. Root MIN has 2 moves left? Actually to reach leaf from root: root (move 0) -> MAX (move 1) -> leaf (move 2). So depth = 2 (internal nodes + leaf? We'll call node_depth = number of levels from current node to leaf. So root has depth 2 (needs two more decisions: its children, then leaves). So we call alpha_beta_min with node_depth=2.
    # However our recursion above assumes when node_depth = 0 we return leaf. So root depth = 2.
    root_depth = 2
    best_value, _, log = alpha_beta_min(root_depth, -math.inf, math.inf, [], 0)
    
    print("=== Traversal Log ===\n")
    print(log)
    
    print("\n=== Results ===")
    print(f"Optimal root value (MIN's best outcome): {best_value}")
    # Determine best move branch for opponent: the MIN node's child that leads to the minimal value
    # We need to re-run to find which branch gave the min value. From log we can see the first child's value? Actually we can compute by evaluating each MAX child's value separately.
    print("\nEvaluating each branch (MAX node values) to find opponent's best move:")
    # Compute each of the three MAX node values (with no pruning to see full values)
    print("\nFull evaluation without pruning (for branch identification):")
    leaf_idx = 0
    branch_values = []
    for b in range(3):
        # Evaluate MAX node with depth 1 (since from MAX to leaf is 1 move)
        # Need a helper that does full eval without pruning? Let's just compute using simple recursion.
        def eval_max_min(node_type, child_list):
            # Simple minimax without pruning to get the actual value of each MAX node.
            # leaf list for this branch: leaf_values[leaf_idx:leaf_idx+3]
            if node_type == 'max':
                return max(eval_min_min([leaf_values[leaf_idx + i] for i in range(3)]))
            else: # min
                return min(child_list)
        # Actually simpler: each MAX node's value is the maximum of its three leaf values.
        values = leaf_values[leaf_idx:leaf_idx+3]
        max_val = max(values)
        branch_values.append(max_val)
        print(f"  Branch {b+1} (MAX node): leaves {values} => value = {max_val}")
        leaf_idx += 3
    print(f"\nBranch values (from left to right): {branch_values}")
    # MIN at root chooses the smallest value among branches
    best_branch_index = branch_values.index(min(branch_values))
    print(f"Opponent (MIN) will choose the branch with smallest value: branch {best_branch_index+1} (value = {branch_values[best_branch_index]})")
    
    print("\n=== Pruned Nodes ===")
    if pruned_nodes:
        for node, reason in pruned_nodes:
            print(f"  Pruned at path {node}: {reason}")
    else:
        print("  No pruning occurred.")
    
    # Visual representation of tree
    print("\n=== Tree Visualization ===")
    print("Root (MIN)")
    for i in range(3):
        print(f"  ├─ MAX node {i+1}")
        for j in range(3):
            leaf_val = leaf_values[i*3 + j]
            print(f"  │   └─ Leaf {j+1}: {leaf_val}")
    print("\nOptimal root value =", best_value)
    print(f"Best first move for opponent (MIN) is to take branch leading to MAX node {best_branch_index+1} (value = {branch_values[best_branch_index]})")

if __name__ == "__main__":
    main()
````

````python
from ortools.sat.python import cp_model

def solve_hospital_scheduling():
    model = cp_model.CpModel()
    
    doctors = ['Aamir', 'Sana', 'Farhan', 'Nadia']
    departments = ['A', 'B', 'C', 'D']  # indices 0,1,2,3
    
    assign = {d: model.NewIntVar(0, 3, d) for d in doctors}
    
    # Each department assigned exactly once
    model.AddAllDifferent([assign[d] for d in doctors])
    
    # Constraint 1: Dr. Aamir must be in Department A (0) or C (2)
    # Create two boolean variables: isA = (assign['Aamir'] == 0), isC = (assign['Aamir'] == 2)
    isA = model.NewBoolVar('Aamir_in_A')
    isC = model.NewBoolVar('Aamir_in_C')
    model.Add(assign['Aamir'] == 0).OnlyEnforceIf(isA)
    model.Add(assign['Aamir'] != 0).OnlyEnforceIf(isA.Not())
    model.Add(assign['Aamir'] == 2).OnlyEnforceIf(isC)
    model.Add(assign['Aamir'] != 2).OnlyEnforceIf(isC.Not())
    model.AddBoolOr([isA, isC])
    
    # Constraint 2: Department C (index 2) requires surgical clearance: only Dr. Farhan or Dr. Aamir
    # So no other doctor can be assigned to C
    for d in doctors:
        if d not in ['Aamir', 'Farhan']:
            model.Add(assign[d] != 2)
    
    # Constraint 3: Department D (index 3) requires child specialist: only Dr. Nadia
    for d in doctors:
        if d != 'Nadia':
            model.Add(assign[d] != 3)
    
    # Dr. Sana cannot be assigned to Department A (index 0)
    model.Add(assign['Sana'] != 0)
    
    # Solve
    solver = cp_model.CpSolver()
    status = solver.Solve(model)
    
    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        print("Solution found:")
        for d in doctors:
            dept_idx = solver.Value(assign[d])
            print(f"{d:6} -> Department {departments[dept_idx]}")
    else:
        print("No solution found.")

if __name__ == "__main__":
    solve_hospital_scheduling()
````

````python
# PAPER Q2 VERSION 2
import numpy as np
import matplotlib.pyplot as plt

# Define state mapping
states = {
    0: "Order Placed",
    1: "Processing",
    2: "Shipped",
    3: "Delivered",
    4: "Returned",
    5: "Cancelled"
}
n_states = len(states)

# Transition matrix
# Rows: current state, columns: next state
# Order: 0:Order Placed, 1:Processing, 2:Shipped, 3:Delivered, 4:Returned, 5:Cancelled
trans_mat = np.zeros((n_states, n_states))

# Order Placed -> Processing (0.95), Cancelled (0.05)
trans_mat[0, 1] = 0.95
trans_mat[0, 5] = 0.05

# Processing -> Shipped (0.7), Processing (0.2), Cancelled (0.1)
trans_mat[1, 2] = 0.7
trans_mat[1, 1] = 0.2
trans_mat[1, 5] = 0.1

# Shipped -> Delivered (0.8), Returned (0.15), Shipped (0.05)
trans_mat[2, 3] = 0.8
trans_mat[2, 4] = 0.15
trans_mat[2, 2] = 0.05

# Delivered -> Delivered (0.95), Returned (0.05)  (return after delivery)
trans_mat[3, 3] = 0.95
trans_mat[3, 4] = 0.05

# Returned -> Returned (absorbing)
trans_mat[4, 4] = 1.0

# Cancelled -> Cancelled (absorbing)
trans_mat[5, 5] = 1.0

# Verify each row sums to 1
print("Transition matrix (rows sum to 1):")
for i in range(n_states):
    print(f"{states[i]}: {trans_mat[i].sum():.2f}")

# Single simulation for 20 steps starting from Order Placed (state 0)
def simulate_single(steps=20, start_state=0):
    state = start_state
    history = [state]
    for _ in range(steps):
        state = np.random.choice(n_states, p=trans_mat[state])
        history.append(state)
    return history

np.random.seed(42)
single_trace = simulate_single(20)
print("\nSingle simulation (20 steps):")
for i, s in enumerate(single_trace):
    print(f"Step {i:2d}: {states[s]}")

# Monte Carlo to estimate probability of Returned before Delivered
def is_returned_before_delivered():
    state = 0  # Order Placed
    while True:
        if state == 4:   # Returned
            return True
        if state == 3:   # Delivered
            return False
        if state == 5:   # Cancelled (neither)
            return False
        # Transition to next state
        state = np.random.choice(n_states, p=trans_mat[state])

n_sim = 10000
np.random.seed(42)
returned_first = 0
for _ in range(n_sim):
    if is_returned_before_delivered():
        returned_first += 1

prob_returned_before_delivered = returned_first / n_sim
print(f"\nMonte Carlo estimate (n={n_sim}):")
print(f"Probability that an order is eventually Returned before Delivered: {prob_returned_before_delivered:.4f}")

# Optional: Show distribution of final states (for verification)
def final_state():
    state = 0
    while True:
        if state in [3,4,5]:
            return state
        state = np.random.choice(n_states, p=trans_mat[state])

final_counts = {3:0, 4:0, 5:0}
for _ in range(n_sim):
    fs = final_state()
    final_counts[fs] += 1
print("\nFinal state distribution (absorbing states):")
print(f"Delivered: {final_counts[3]/n_sim:.3f}")
print(f"Returned:   {final_counts[4]/n_sim:.3f}")
print(f"Cancelled:  {final_counts[5]/n_sim:.3f}")
````

````python
# PAPER QUESTION 3 VERSION 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ------------------------------
# Generate synthetic house price dataset (since no file provided)
# ------------------------------
np.random.seed(42)
n = 1000

data = {
    'HouseID': range(1, n+1),
    'Area_sqft': np.random.normal(2000, 500, n).clip(800, 5000),
    'Bedrooms': np.random.choice([1,2,3,4,5], n, p=[0.05,0.2,0.4,0.25,0.1]),
    'Bathrooms': np.random.choice([1,2,3,4], n, p=[0.1,0.5,0.3,0.1]),
    'Age_years': np.random.exponential(20, n).clip(0, 100),
    'Location_Score': np.random.uniform(1, 10, n),  # 1=worst,10=best
    'Garage_Spaces': np.random.choice([0,1,2,3], n, p=[0.2,0.5,0.25,0.05]),
    'Condition_Score': np.random.uniform(1, 10, n),
    'SalePrice': 0
}
df = pd.DataFrame(data)

# Generate SalePrice = base + area*coef + bedrooms*coef + bathrooms*coef - age*coef + location*coef + noise
df['SalePrice'] = (
    50000 +
    df['Area_sqft'] * 120 +
    df['Bedrooms'] * 8000 +
    df['Bathrooms'] * 10000 -
    df['Age_years'] * 1500 +
    df['Location_Score'] * 8000 +
    df['Garage_Spaces'] * 5000 +
    df['Condition_Score'] * 3000 +
    np.random.normal(0, 20000, n)
).clip(50000, 800000)

# Introduce intentional missing values for demonstration
missing_idx = np.random.choice(n, 30, replace=False)
df.loc[missing_idx, 'Area_sqft'] = np.nan
missing_idx2 = np.random.choice(n, 20, replace=False)
df.loc[missing_idx2, 'Bathrooms'] = np.nan

print("=== PART 1: DATA PREPROCESSING ===\n")

# ------------------------------
# 1. Dimensions, data types, statistical summary
# ------------------------------
print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("\nData types:\n", df.dtypes)
print("\nStatistical summary (numerical columns):")
print(df.describe())

# ------------------------------
# 2. Missing values and median imputation
# ------------------------------
print("\nMissing values before imputation:")
print(df.isnull().sum())
for col in df.columns:
    if df[col].dtype in ['float64', 'int64'] and df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
print("\nMissing values after imputation:")
print(df.isnull().sum())

# ------------------------------
# 3. EDA: histogram of SalePrice, correlation heatmap, top features
# ------------------------------
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['SalePrice'], kde=True, bins=30)
plt.title('Distribution of SalePrice')

# Correlation heatmap
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr = df[numeric_cols].corr()
plt.subplot(1,2,2)
sns.heatmap(corr, annot=False, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

# Top 3 features correlated with SalePrice (excluding SalePrice itself)
corr_with_target = corr['SalePrice'].drop('SalePrice').abs().sort_values(ascending=False)
top3 = corr_with_target.head(3).index.tolist()
print(f"\nTop 3 features most correlated with SalePrice: {top3}")

# Scatter plots for SalePrice vs top features
fig, axes = plt.subplots(1, 3, figsize=(15,4))
for i, feat in enumerate(top3):
    axes[i].scatter(df[feat], df['SalePrice'], alpha=0.5)
    axes[i].set_xlabel(feat)
    axes[i].set_ylabel('SalePrice')
    axes[i].set_title(f'{feat} vs SalePrice (corr={corr.loc[feat,"SalePrice"]:.2f})')
plt.tight_layout()
plt.show()

# Box plots for SalePrice and key features to detect outliers
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
sns.boxplot(y=df['SalePrice'])
plt.title('SalePrice Boxplot')
plt.subplot(1,3,2)
sns.boxplot(y=df['Area_sqft'])
plt.title('Area_sqft Boxplot')
plt.subplot(1,3,3)
sns.boxplot(y=df['Age_years'])
plt.title('Age_years Boxplot')
plt.tight_layout()
plt.show()
print("Note: Some extreme SalePrice values (>600k) appear as mild outliers, but they are plausible.")

# ------------------------------
# 4. StandardScaler, drop ID, separate X and y
# ------------------------------
df_clean = df.drop('HouseID', axis=1)
X = df_clean.drop('SalePrice', axis=1)
y = df_clean['SalePrice']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print("\nFeature matrix X shape:", X_scaled.shape)
print("Target vector y shape:", y.shape)

# ------------------------------
# PART 2: MODEL TRAINING
# ------------------------------
print("\n=== PART 2: MODEL TRAINING ===\n")

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

# 1. KNN Regressor with k=5, Euclidean (default metric)
knn = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)

# 2. Decision Tree Regressor with max_depth=5, min_samples_split=10, min_samples_leaf=5
dt = DecisionTreeRegressor(max_depth=5, min_samples_split=10, min_samples_leaf=5, random_state=42)
dt.fit(X_train, y_train)

# 3. Linear Regression with fit_intercept=True
lr = LinearRegression(fit_intercept=True)
lr.fit(X_train, y_train)

models = {'KNN': knn, 'Decision Tree': dt, 'Linear Regression': lr}

# ------------------------------
# PART 3: MODEL EVALUATION
# ------------------------------
print("=== PART 3: MODEL EVALUATION ===\n")

def evaluate_model(model, X, y, set_name=""):
    pred = model.predict(X)
    mae = mean_absolute_error(y, pred)
    mse = mean_squared_error(y, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, pred)
    return pd.Series({'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2})

# Training set evaluation
print("Training Set Performance:")
train_results = {}
for name, model in models.items():
    train_results[name] = evaluate_model(model, X_train, y_train, "Train")
train_df = pd.DataFrame(train_results).T
print(train_df.round(4))

# Test set evaluation
print("\nTest Set Performance:")
test_results = {}
for name, model in models.items():
    test_results[name] = evaluate_model(model, X_test, y_test, "Test")
test_df = pd.DataFrame(test_results).T
print(test_df.round(4))

# Confusion matrix is not applicable for regression; but we can show residual distribution
print("\nNote: Confusion matrix is for classification. For regression we use residual plots.")
best_model_name = test_df['R2'].idxmax()
best_model = models[best_model_name]
print(f"\nBest model on test set (by R2): {best_model_name}")

# Actual vs Predicted scatter plot for best model
y_pred_best = best_model.predict(X_test)
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_best, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual SalePrice')
plt.ylabel('Predicted SalePrice')
plt.title(f'{best_model_name}: Actual vs Predicted (Test Set)')
plt.tight_layout()
plt.show()

# 10-fold cross-validation
print("\n10-Fold Cross-Validation Results (Average R2):")
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
cv_results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_scaled, y, cv=kfold, scoring='r2')
    cv_results[name] = (scores.mean(), scores.std())
    print(f"{name:20} Mean R2 = {scores.mean():.4f} (+/- {scores.std():.4f})")

# ------------------------------
# CONCLUSION
# ------------------------------
print("\n=== CONCLUSION ===")
print("Based on test set R² and RMSE, the best model is", best_model_name)
print("Linear Regression likely generalizes best; Decision Tree may slightly underfit due to depth limit; KNN performance depends on scale (already scaled).")
print("None of the models show severe overfitting (train vs test R² close).")
print("A real-world limitation: our model assumes linear relationships and independent features, but house prices are influenced by many unobserved factors (e.g., school quality, economic trends).")
print("Additionally, the synthetic data may not capture the complexity of real estate markets.")
````

````python
from ortools.sat.python import cp_model

def solve_vehicle_assignment():
    model = cp_model.CpModel()
    
    locations = ['L1', 'L2', 'L3', 'L4', 'L5']
    vehicle = {loc: model.NewIntVar(0, 2, loc) for loc in locations}
    
    # Capacity variables
    count_v1 = model.NewIntVar(0, 2, "count_v1")
    count_v2 = model.NewIntVar(0, 2, "count_v2")
    count_v3 = model.NewIntVar(0, 3, "count_v3")
    
    # Create indicator variables for each location-vehicle assignment
    for loc in locations:
        is_v1 = model.NewBoolVar(f"{loc}_v1")
        is_v2 = model.NewBoolVar(f"{loc}_v2")
        is_v3 = model.NewBoolVar(f"{loc}_v3")
        model.Add(vehicle[loc] == 0).OnlyEnforceIf(is_v1)
        model.Add(vehicle[loc] == 1).OnlyEnforceIf(is_v2)
        model.Add(vehicle[loc] == 2).OnlyEnforceIf(is_v3)
        model.Add(vehicle[loc] != 0).OnlyEnforceIf(is_v1.Not())
        model.Add(vehicle[loc] != 1).OnlyEnforceIf(is_v2.Not())
        model.Add(vehicle[loc] != 2).OnlyEnforceIf(is_v3.Not())
        
        # Add to counts using linear expressions
        model.Add(count_v1 == count_v1 + is_v1)  # will be summed differently
        # Instead, we'll sum after loops
    
    # Better: sum indicators directly
    sum_v1 = sum(model.NewBoolVar(f"{loc}_v1") for loc in locations)
    sum_v2 = sum(model.NewBoolVar(f"{loc}_v2") for loc in locations)
    sum_v3 = sum(model.NewBoolVar(f"{loc}_v3") for loc in locations)
    
    # Recreate indicators properly
    # Clear previous and do correctly:
    # We'll recreate the section
    # Actually simpler: use model.Add(count_v1 == sum(vehicle[loc] == 0)) requires AddBoolOr trick?
    # Use LinearExpr.Sum with BooleanVar objects:
    from ortools.sat.python import cp_model_helper  # not needed; use built-in
    # Correct approach:
    indicators = {v: [] for v in range(3)}
    for loc in locations:
        for v in range(3):
            ind = model.NewBoolVar(f"{loc}_v{v}")
            model.Add(vehicle[loc] == v).OnlyEnforceIf(ind)
            model.Add(vehicle[loc] != v).OnlyEnforceIf(ind.Not())
            indicators[v].append(ind)
    
    model.Add(count_v1 == sum(indicators[0]))
    model.Add(count_v2 == sum(indicators[1]))
    model.Add(count_v3 == sum(indicators[2]))
    
    # L1 and L2 different vehicle
    model.Add(vehicle['L1'] != vehicle['L2'])
    # L3 must be V3
    model.Add(vehicle['L3'] == 2)
    
    # Workload difference between V1 and V2 <= 1
    diff = model.NewIntVar(-2, 2, "diff")
    model.Add(diff == count_v1 - count_v2)
    model.Add(diff >= -1)
    model.Add(diff <= 1)
    
    # Dependency: if L4 is V1 then L5 is V1
    l4_is_v1 = model.NewBoolVar("l4_v1")
    model.Add(vehicle['L4'] == 0).OnlyEnforceIf(l4_is_v1)
    model.Add(vehicle['L4'] != 0).OnlyEnforceIf(l4_is_v1.Not())
    # Implication
    l5_is_v1 = model.NewBoolVar("l5_v1")
    model.Add(vehicle['L5'] == 0).OnlyEnforceIf(l5_is_v1)
    model.Add(vehicle['L5'] != 0).OnlyEnforceIf(l5_is_v1.Not())
    model.AddImplication(l4_is_v1, l5_is_v1)
    
    solver = cp_model.CpSolver()
    status = solver.Solve(model)
    
    if status in (cp_model.OPTIMAL, cp_model.FEASIBLE):
        print("Valid vehicle assignment:")
        for loc in locations:
            v = solver.Value(vehicle[loc])
            print(f"{loc} -> V{v+1}")
        print(f"\nDelivery counts: V1={solver.Value(count_v1)}, V2={solver.Value(count_v2)}, V3={solver.Value(count_v3)}")
    else:
        print("No solution found.")

if __name__ == "__main__":
    solve_vehicle_assignment()
````

BAYESIAN DISEASE PROBLEM

````python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Define network structure
model = DiscreteBayesianNetwork([
    ('Disease', 'Fever'),
    ('Disease', 'Cough'),
    ('Disease', 'Fatigue'),
    ('Disease', 'Chills')
])

# Prior for Disease: Flu, Cold
cpd_disease = TabularCPD('Disease', 2, [[0.3], [0.7]],
                         state_names={'Disease': ['Flu', 'Cold']})

# CPTs for symptoms (each: 2 states: Yes/No, evidence: Disease)
# Fever
fever_cpt = [[0.1, 0.5],   # Fever=No given Flu, Cold
             [0.9, 0.5]]   # Fever=Yes given Flu, Cold
cpd_fever = TabularCPD('Fever', 2, fever_cpt,
                       evidence=['Disease'], evidence_card=[2],
                       state_names={'Fever': ['No','Yes'], 'Disease': ['Flu','Cold']})

# Cough
cough_cpt = [[0.2, 0.4],   # Cough=No
             [0.8, 0.6]]   # Cough=Yes
cpd_cough = TabularCPD('Cough', 2, cough_cpt,
                       evidence=['Disease'], evidence_card=[2],
                       state_names={'Cough': ['No','Yes'], 'Disease': ['Flu','Cold']})

# Fatigue
fatigue_cpt = [[0.3, 0.7],   # Fatigue=No
               [0.7, 0.3]]   # Fatigue=Yes
cpd_fatigue = TabularCPD('Fatigue', 2, fatigue_cpt,
                         evidence=['Disease'], evidence_card=[2],
                         state_names={'Fatigue': ['No','Yes'], 'Disease': ['Flu','Cold']})

# Chills
chills_cpt = [[0.4, 0.6],   # Chills=No
              [0.6, 0.4]]   # Chills=Yes
cpd_chills = TabularCPD('Chills', 2, chills_cpt,
                        evidence=['Disease'], evidence_card=[2],
                        state_names={'Chills': ['No','Yes'], 'Disease': ['Flu','Cold']})

# Add all CPDs
model.add_cpds(cpd_disease, cpd_fever, cpd_cough, cpd_fatigue, cpd_chills)
assert model.check_model()

inference = VariableElimination(model)

# Inference Task 1: P(Disease | Fever=Yes, Cough=Yes)
result1 = inference.query(variables=['Disease'], 
                          evidence={'Fever': 'Yes', 'Cough': 'Yes'})
print("P(Disease | Fever=Yes, Cough=Yes):")
print(result1)

# Inference Task 2: P(Disease | Fever=Yes, Cough=Yes, Chills=Yes)
result2 = inference.query(variables=['Disease'],
                          evidence={'Fever': 'Yes', 'Cough': 'Yes', 'Chills': 'Yes'})
print("\nP(Disease | Fever=Yes, Cough=Yes, Chills=Yes):")
print(result2)

# Inference Task 3: P(Fatigue=Yes | Disease=Flu)
# This is directly from CPT, but we can also query
result3 = inference.query(variables=['Fatigue'], evidence={'Disease': 'Flu'})
print("\nP(Fatigue=Yes | Disease=Flu):")
print(result3)
````
